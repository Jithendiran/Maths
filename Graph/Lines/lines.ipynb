| Geometric Figure | Description | Notation | Extends Infinitely? |
| :--- | :--- | :--- | :--- |
| **Line** | A straight path that extends forever in **both** directions. | $\overleftrightarrow{AB}$ | Yes, in two directions |
| **Ray** | A part of a line that starts at one endpoint and extends forever in **one** direction. | $\overrightarrow{AB}$ | Yes, in one direction |
| **Segment** | A part of a line with **two** distinct endpoints. | $\overline{AB}$ | **No** (has a finite length) |

```python
import matplotlib.pyplot as plt
import numpy as np

# Create x values for the lines
x = np.linspace(-5, 5, 100)

# Define the three parallel lines (m=1, different intercepts)
y1 = x           # y = x + 0
y2 = x + 3       # y = x + 3
y3 = x - 2       # y = x - 2

# Create the plot
plt.figure(figsize=(7, 7))

# Plot the lines
plt.plot(x, y1, label='Line 1 (m=1, b=0)', color='blue')
plt.plot(x, y2, label='Line 2 (m=1, b=3)', linestyle='--', color='red')
plt.plot(x, y3, label='Line 3 (m=1, b=-2)', linestyle=':', color='green')

# Set the aspect ratio to 'equal' so the lines look correctly angled
plt.gca().set_aspect('equal', adjustable='box')

# Add labels, title, and grid
plt.title('Parallel Lines (Same Slope)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axhline(0, color='black', linewidth=0.5) # x-axis
plt.axvline(0, color='black', linewidth=0.5) # y-axis
plt.grid(True, linestyle='dotted', alpha=0.6)
plt.legend()
plt.show()
```

```python
# Create x values for the lines
x = np.linspace(-5, 5, 100)

# Define the two perpendicular lines
# Line A: m=1, b=1
y_a = x + 1
# Line B: m=-1, b=1 (Negative Reciprocal)
y_b = -x + 1

# Create the plot
plt.figure(figsize=(7, 7))

# Plot the lines
plt.plot(x, y_a, label='Line A (Slope = 1)', color='blue', linewidth=2)
plt.plot(x, y_b, label='Line B (Slope = -1)', color='red', linewidth=2)

# Mark the intersection point (1, 1) and the y-intercept (0, 1)
plt.scatter([0], [1], color='black', s=50, zorder=5) # Intersection point (0, 1)

# Set the aspect ratio to 'equal' so the 90-degree angle looks correct
plt.gca().set_aspect('equal', adjustable='box')

# Add labels, title, and grid
plt.title('Perpendicular Lines (Slopes: 1 and -1)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='dotted', alpha=0.6)
plt.legend()
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.show()
```
