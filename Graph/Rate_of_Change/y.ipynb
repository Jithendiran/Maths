# Slope

[Read Rate of change](../Knows/Division.md#rate-of-change)


$$\frac{\Delta y}{\Delta x} = \text{Constant Rate of }\Delta y \text{ for single unit of } \Delta x$$
Here y denotes the up or down change of vertical line (y-axis) and x denotes the left or right of the horizontal line (x-axis)
What we find here is for 1 unit of x value, how much up or down the y value is 

if x moves (left or right) 1 unit of point, y will definitely move (up or down)
$\frac{\Delta y}{\Delta x}$ will give you constant value for rate of change of y

Here we are getting output of steepness (How the y moves up or down based on any value of x)

**Steepness** is a general term that describes how sharply a surface, such as a slope or a hill, ascends (goes up) or descends (goes down).

**Slope** is a measurement of a line's steepness and direction


$$Slope = \frac{y2-y1}{x2-x1} = \frac{\Delta y}{\Delta x}$$
In algebra slope is denoted in $m$


P1(0,0), P2(0,0) -> (0)/(0) moving no where

P1(0,0), P2(1,1) -> 1/1  = 1 moving right, up

P1(0,0), P2(-1,1) -> -1/1 = -1 moving right, down

P1(0,0), P2(1,-1) -> 1/-1 = -1 moving left, up

P1(0,0), P2(-1,-1) -> -1/-1 = 1 moving left, down

P1(-1,-1), P2(1,1) -> ((-1)-(1))/((-1)-(1)) = -2/-2 = 1

## Positive qudarent

$$m = \frac{y_2 - y_1}{x_2 - x_1}$$

### Positive steep

If $y_2 > y_1$, the $\text{Rise}$ (numerator) is positive.  

If $x_2 > x_1$, the $\text{Run}$ (denominator) is positive.  

A positive number divided by a positive number yields a positive slope.

1. $m=0$ (Horizontal): Points $(0, 0)$ and $(4, 0)$.

    $$m = \frac{0 - 0}{4 - 0} = \frac{0}{4} = \mathbf{0}$$

2. $m=0.5$ (Shallow): Points $(0, 0)$ and $(4, 2)$.

    $$m = \frac{2 - 0}{4 - 0} = \frac{2}{4} = \frac{1}{2} = \mathbf{0.5}$$

3. $m=1$ (Moderate): Points $(0, 0)$ and $(4, 4)$.

    $$m = \frac{4 - 0}{4 - 0} = \frac{4}{4} = \mathbf{1}$$

4. $m=2$ (Steep): Points $(0, 0)$ and $(2, 4)$.

    $$m = \frac{4 - 0}{2 - 0} = \frac{4}{2} = \mathbf{2}$$

5. $m=\infty$ (Vertical): Points $(1, 0)$ and $(1, 4)$

    $$m = \frac{4 - 0}{1 - 1} = \frac{4}{0} = \infty$$

### negative steep

If $y_2 < y_1$, the $\text{Rise}$ (numerator) is negative.

If $x_2 > x_1$, the $\text{Run}$ (denominator) is positive.

A negative number divided by a positive number yields a negative slope.

6. $m=-0.5$ (Shallow): Points $(4, 4)$ and $(8, 2)$

    $$m = \frac{2 - 4}{8 - 4} = \frac{-2}{4} = \mathbf{-0.5}$$

7. $m=-1$ (Shallow): Points $(4, 4)$ and $(8, 0)$

    $$m = \frac{0 - 4}{8 - 4} = \frac{-4}{4} = \mathbf{-1}$$

8. $m=-2$ (Steep): Points $(4, 4)$ and $(6, 0)$.

    $$m = \frac{0 - 4}{6 - 4} = \frac{-4}{2} = \mathbf{-2}$$

Horizontal and Vertical are same as above

**Positive slope**: As the x-value increases, the y-value also increases.
Analogy: Think of climbing a hill. As you move forward (increasing x), you go up (increasing y).

**Negative slope**: As the x-value increases, the y-value decreases.
Analogy: Think of sliding down a slide. As you move horizontally forward (increasing x), your height decreases (decreasing y).

### Moving left

The definition of positive and negative slope is based on the traditional reading direction: moving from left to right (where the x-value is increasing).


#### Falling
P1 = (4,4) and p2 = (1,3)
$$m = \frac{3-4}{1-4}=\frac{-1}{-3}=\frac{1}{3}=0.3$$

as per m, point moving upwords but actually point is moved downwards

#### Raising
p1 = (4,4) and p2 = (1, 5)

$$m = \frac{5-4}{1-4}=\frac{1}{-3}=\frac{1}{-3}=-0.3$$

as per m, point moving downwards but actually point is moved upwards


```python
# Set up the figure and axes
fig, ax = plt.subplots(figsize=(10, 8))

# Set the title and labels
ax.set_title('Comparison of Slopes: m = 0, 0.5, 1, 2, and $\\pm\\infty$', fontsize=16, pad=20)
ax.set_xlabel(r'$x$ (Run)')
ax.set_ylabel(r'$y$ (Rise)')
ax.grid(True, linestyle='--', alpha=0.6)

# Set the limits to focus on the first quadrant and include the vertical line
plot_limit = 5.0
ax.set_xlim(-0.5, plot_limit)
ax.set_ylim(-0.5, plot_limit)
ax.set_aspect('equal', adjustable='box')

# Draw the axes
ax.axhline(0, color='black', linewidth=1.5, zorder=1)
ax.axvline(0, color='black', linewidth=1.5, zorder=1)

# Generate x-values for diagonal and horizontal lines (starting at x=0)
x_line = np.linspace(0, plot_limit, 100)

# --- Define a reference x-value for marking points (The "Run") ---
x_mark = 4.0

# --- Plot m = 2 (Steep Positive) ---
m2 = 2
line2, = ax.plot(x_line, m2 * x_line, color='darkgreen', linewidth=3, alpha=0.8, label=r'$m=2$ (Steep)')
# Mark the point (Note: This point (4, 8) is outside the set y-limit of 5.0, so it won't be visible)
# To show the steepness of m=2 within the view, we'll mark a point at a smaller x-value, say x=2.
x_mark_steep = 2.0
y_mark_steep = m2 * x_mark_steep
ax.plot(x_mark_steep, y_mark_steep, 'o', color='darkgreen', markersize=8, zorder=3)
ax.text(x_mark_steep, y_mark_steep + 0.15, f'({x_mark_steep:.0f}, {y_mark_steep:.0f})', color='darkgreen', fontsize=10, ha='center', va='bottom')


# --- Plot m = 1 (Moderate Positive) ---
m1 = 1
line1, = ax.plot(x_line, m1 * x_line, color='blue', linewidth=3, alpha=0.8, label=r'$m=1$ (Moderate)')
# Mark the point
y_mark_m1 = m1 * x_mark
ax.plot(x_mark, y_mark_m1, 'o', color='blue', markersize=8, zorder=3)
ax.text(x_mark, y_mark_m1 + 0.15, f'({x_mark:.0f}, {y_mark_m1:.0f})', color='blue', fontsize=10, ha='center', va='bottom')

# --- Plot m = 0.5 (Shallow Positive) ---
m05 = 0.5
line05, = ax.plot(x_line, m05 * x_line, color='darkorange', linewidth=3, alpha=0.8, label=r'$m=0.5$ (Shallow)')
# Mark the point
y_mark_m05 = m05 * x_mark
ax.plot(x_mark, y_mark_m05, 'o', color='darkorange', markersize=8, zorder=3)
ax.text(x_mark, y_mark_m05 + 0.15, f'({x_mark:.0f}, {y_mark_m05:.1f})', color='darkorange', fontsize=10, ha='center', va='bottom')

# --- Plot m = 0 (Horizontal) ---
m0 = 0
line0, = ax.plot(x_line, m0 * x_line, color='darkred', linestyle='-', linewidth=3, alpha=0.8, label=r'$m=0$ (Horizontal)')
# Mark the point
y_mark_m0 = m0 * x_mark
ax.plot(x_mark, y_mark_m0, 'o', color='darkred', markersize=8, zorder=3)
ax.text(x_mark, y_mark_m0 - 0.15, f'({x_mark:.0f}, {y_mark_m0:.0f})', color='darkred', fontsize=10, ha='center', va='top')

# --- VISUAL EXAMPLE: Rise over Run for m=1 ---
run_end = x_mark
rise_end = y_mark_m1
# Draw the 'Run' line (Horizontal)
ax.plot([0, run_end], [0, 0], color='blue', linestyle=':', linewidth=1.5, alpha=0.7)
ax.plot([0, run_end], [rise_end, rise_end], color='blue', linestyle=':', linewidth=1.5, alpha=0.7)
# Draw the 'Rise' line (Vertical)
ax.plot([run_end, run_end], [0, rise_end], color='blue', linestyle=':', linewidth=1.5, alpha=0.7)



# --- Plot m = +inf (Vertical) ---
# The vertical line (x=constant) must be plotted separately. We use x=1 to place it clearly.
x_val_vert = 1.0
y_line_vert = np.linspace(0, plot_limit, 100)
x_line_vert = np.full_like(y_line_vert, x_val_vert)
ax.plot(x_line_vert, y_line_vert, color='gray', linestyle='--', linewidth=3, label=r'$x=1$ ($m=\pm\infty$, Vertical)')
ax.text(x_val_vert - 0.15, plot_limit * 0.8, r'No Run ($m=\pm\infty$)', color='gray', fontsize=10, ha='right', va='center', rotation=90)


# --- Annotation for comparison (Moved to be clearly inside the plot) ---
# Slope definition moved up to avoid the lines, but still inside the y=5 boundary
ax.text(0.5, 4.8, r'$\text{Slope } (m) = \frac{\Delta y}{\Delta x} = \frac{\text{Rise}}{\text{Run}}$',
        color='black', fontsize=14, ha='center', va='center',
        bbox=dict(boxstyle="round,pad=0.4", fc="lightblue", alpha=0.8))



# --- Legend (Moved to the top center) ---
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05), ncol=5, fontsize='medium')

plt.show()
```

```python
# Set up the figure and axes
fig, ax = plt.subplots(figsize=(10, 8))

# Set the title and labels
ax.set_title('Comparison of Slopes: Positive, Negative, and Extreme', fontsize=16, pad=20)
ax.set_xlabel(r'$x$ (Run)')
ax.set_ylabel(r'$y$ (Rise)')
ax.grid(True, linestyle='--', alpha=0.6)

# Set the limits to focus on the first quadrant and include the vertical line
plot_limit = 5.0
ax.set_xlim(-0.5, plot_limit)
ax.set_ylim(-0.5, plot_limit)
ax.set_aspect('equal', adjustable='box')

# Draw the axes
ax.axhline(0, color='black', linewidth=1.5, zorder=1)
ax.axvline(0, color='black', linewidth=1.5, zorder=1)

# Generate x-values for diagonal and horizontal lines (starting at x=0)
x_line = np.linspace(0, plot_limit, 100)

# --- Define a reference x-value for marking points (The "Run") ---
x_mark = 4.0

# --- Positive Slopes (Using the y=mx form, starting at the origin) ---

# --- Plot m = 2 (Steep Positive) ---
m2 = 2
line2, = ax.plot(x_line, m2 * x_line, color='darkgreen', linewidth=3, alpha=0.8, label=r'$m=2$ (Steep +)')
x_mark_steep = 2.0
y_mark_steep = m2 * x_mark_steep
ax.plot(x_mark_steep, y_mark_steep, 'o', color='darkgreen', markersize=8, zorder=3)
ax.text(x_mark_steep, y_mark_steep + 0.15, f'({x_mark_steep:.0f}, {y_mark_steep:.0f})', color='darkgreen', fontsize=10, ha='center', va='bottom')


# --- Plot m = 1 (Moderate Positive) ---
m1 = 1
line1, = ax.plot(x_line, m1 * x_line, color='blue', linewidth=3, alpha=0.8, label=r'$m=1$ (Moderate +)')
y_mark_m1 = m1 * x_mark
ax.plot(x_mark, y_mark_m1, 'o', color='blue', markersize=8, zorder=3)
ax.text(x_mark, y_mark_m1 + 0.15, f'({x_mark:.0f}, {y_mark_m1:.0f})', color='blue', fontsize=10, ha='center', va='bottom')

# --- Plot m = 0.5 (Shallow Positive) ---
m05 = 0.5
line05, = ax.plot(x_line, m05 * x_line, color='darkorange', linewidth=3, alpha=0.8, label=r'$m=0.5$ (Shallow +)')
y_mark_m05 = m05 * x_mark
ax.plot(x_mark, y_mark_m05, 'o', color='darkorange', markersize=8, zorder=3)
ax.text(x_mark, y_mark_m05 + 0.15, f'({x_mark:.0f}, {y_mark_m05:.1f})', color='darkorange', fontsize=10, ha='center', va='bottom')

# --- Plot m = 0 (Horizontal) ---
m0 = 0
line0, = ax.plot(x_line, m0 * x_line, color='darkred', linestyle='-', linewidth=3, alpha=0.8, label=r'$m=0$ (Horizontal)')
y_mark_m0 = m0 * x_mark
ax.plot(x_mark, y_mark_m0, 'o', color='darkred', markersize=8, zorder=3)
ax.text(x_mark, y_mark_m0 - 0.15, f'({x_mark:.0f}, {y_mark_m0:.0f})', color='darkred', fontsize=10, ha='center', va='top')

# --- VISUAL EXAMPLE: Rise over Run for m=1 ---
run_end = x_mark
rise_end = y_mark_m1
# Draw the 'Run' line (Horizontal)
ax.plot([0, run_end], [0, 0], color='blue', linestyle=':', linewidth=1.5, alpha=0.7)
ax.plot([0, run_end], [rise_end, rise_end], color='blue', linestyle=':', linewidth=1.5, alpha=0.7)
# Draw the 'Rise' line (Vertical)
ax.plot([run_end, run_end], [0, rise_end], color='blue', linestyle=':', linewidth=1.5, alpha=0.7)


# ----------------------------------------------------------------------
# ### NEGATIVE SLOPES (Must use y = mx + b to keep them in the 1st Quadrant) ###
# ----------------------------------------------------------------------
y_int = 5.0 # Set a positive y-intercept so the lines pass through Quadrant I

# --- Plot m = -2 (Steep Negative) ---
m_n2 = -2
line_n2, = ax.plot(x_line, m_n2 * x_line + y_int, color='purple', linestyle='-', linewidth=3, alpha=0.8, label=r'$m=-2$ (Steep -)')
# Mark a point on the line (x=2)
x_mark_n2 = 2.0
y_mark_n2 = m_n2 * x_mark_n2 + y_int
ax.plot(x_mark_n2, y_mark_n2, 's', color='purple', markersize=8, zorder=3)
ax.text(x_mark_n2 + 0.15, y_mark_n2, f'({x_mark_n2:.0f}, {y_mark_n2:.0f})', color='purple', fontsize=10, ha='left', va='center')


# --- Plot m = -1 (Moderate Negative) ---
m_n1 = -1
line_n1, = ax.plot(x_line, m_n1 * x_line + y_int, color='firebrick', linestyle='-', linewidth=3, alpha=0.8, label=r'$m=-1$ (Moderate -)')
# Mark a point on the line (x=3)
x_mark_n1 = 3.0
y_mark_n1 = m_n1 * x_mark_n1 + y_int
ax.plot(x_mark_n1, y_mark_n1, 's', color='firebrick', markersize=8, zorder=3)
ax.text(x_mark_n1 + 0.15, y_mark_n1, f'({x_mark_n1:.0f}, {y_mark_n1:.0f})', color='firebrick', fontsize=10, ha='left', va='center')


# --- Plot m = -0.5 (Shallow Negative) ---
m_n05 = -0.5
line_n05, = ax.plot(x_line, m_n05 * x_line + y_int, color='darkcyan', linestyle='-', linewidth=3, alpha=0.8, label=r'$m=-0.5$ (Shallow -)')
# Mark a point on the line (x=4)
x_mark_n05 = 4.0
y_mark_n05 = m_n05 * x_mark_n05 + y_int
ax.plot(x_mark_n05, y_mark_n05, 's', color='darkcyan', markersize=8, zorder=3)
ax.text(x_mark_n05 + 0.15, y_mark_n05, f'({x_mark_n05:.0f}, {y_mark_n05:.0f})', color='darkcyan', fontsize=10, ha='left', va='center')


# --- Plot m = +inf (Vertical) ---
# The vertical line (x=constant) must be plotted separately. We use x=1 to place it clearly.
x_val_vert = 1.0
y_line_vert = np.linspace(0, plot_limit, 100)
x_line_vert = np.full_like(y_line_vert, x_val_vert)
ax.plot(x_line_vert, y_line_vert, color='gray', linestyle='--', linewidth=3, label=r'$x=1$ ($m=\pm\infty$, Vertical)')
ax.text(x_val_vert - 0.15, plot_limit * 0.8, r'No Run ($m=\pm\infty$)', color='gray', fontsize=10, ha='right', va='center', rotation=90)


# --- Annotation for comparison ---
ax.text(0.5, 4.8, r'$\text{Slope } (m) = \frac{\Delta y}{\Delta x} = \frac{\text{Rise}}{\text{Run}}$',
        color='black', fontsize=14, ha='center', va='center',
        bbox=dict(boxstyle="round,pad=0.4", fc="lightblue", alpha=0.8))


# --- Legend ---
# Adjusted location to fit the new lines
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05), ncol=4, fontsize='small')

plt.show()
```
