# Slope

[Read Rate of change](../../Knows/Division.md#rate-of-change)


$$\frac{\Delta y}{\Delta x} = \text{Constant Rate of }\Delta y \text{ for single unit of } \Delta x$$
Here y denotes the up or down change of vertical line (y-axis) and x denotes the left or right of the horizontal line (x-axis)
What we find here is for 1 unit of x value, how much up or down the y value is 

if x moves (left or right) 1 unit of point, y will definitely move (up or down)
$\frac{\Delta y}{\Delta x}$ will give you constant value for rate of change of y

Here we are getting output of steepness (How the y moves up or down based on any value of x)

**Steepness** is a general term that describes how sharply a surface, such as a slope or a hill, ascends (goes up) or descends (goes down).

**Slope** is a measurement of a line's steepness and direction


$$Slope = \frac{y2-y1}{x2-x1} = \frac{\Delta y}{\Delta x}$$
In algebra slope is denoted in $m$


P1(0,0), P2(0,0) -> (0)/(0) moving no where

P1(0,0), P2(1,1) -> 1/1  = 1 moving right, up

P1(0,0), P2(-1,1) -> -1/1 = -1 moving right, down

P1(0,0), P2(1,-1) -> 1/-1 = -1 moving left, up

P1(0,0), P2(-1,-1) -> -1/-1 = 1 moving left, down

P1(-1,-1), P2(1,1) -> ((-1)-(1))/((-1)-(1)) = -2/-2 = 1

## Positive qudarent

$$m = \frac{y_2 - y_1}{x_2 - x_1}$$

### Positive steep

If $y_2 > y_1$, the $\text{Rise}$ (numerator) is positive.  

If $x_2 > x_1$, the $\text{Run}$ (denominator) is positive.  

A positive number divided by a positive number yields a positive slope.

1. $m=0$ (Horizontal): Points $(0, 0)$ and $(4, 0)$.

    $$m = \frac{0 - 0}{4 - 0} = \frac{0}{4} = \mathbf{0}$$

2. $m=0.5$ (Shallow): Points $(0, 0)$ and $(4, 2)$.

    $$m = \frac{2 - 0}{4 - 0} = \frac{2}{4} = \frac{1}{2} = \mathbf{0.5}$$

3. $m=1$ (Moderate): Points $(0, 0)$ and $(4, 4)$.

    $$m = \frac{4 - 0}{4 - 0} = \frac{4}{4} = \mathbf{1}$$

4. $m=2$ (Steep): Points $(0, 0)$ and $(2, 4)$.

    $$m = \frac{4 - 0}{2 - 0} = \frac{4}{2} = \mathbf{2}$$

5. $m=\infty$ (Vertical): Points $(1, 0)$ and $(1, 4)$

    $$m = \frac{4 - 0}{1 - 1} = \frac{4}{0} = \infty$$

### negative steep

If $y_2 < y_1$, the $\text{Rise}$ (numerator) is negative.

If $x_2 > x_1$, the $\text{Run}$ (denominator) is positive.

A negative number divided by a positive number yields a negative slope.

6. $m=-0.5$ (Shallow): Points $(4, 4)$ and $(8, 2)$

    $$m = \frac{2 - 4}{8 - 4} = \frac{-2}{4} = \mathbf{-0.5}$$

7. $m=-1$ (Shallow): Points $(4, 4)$ and $(8, 0)$

    $$m = \frac{0 - 4}{8 - 4} = \frac{-4}{4} = \mathbf{-1}$$

8. $m=-2$ (Steep): Points $(4, 4)$ and $(6, 0)$.

    $$m = \frac{0 - 4}{6 - 4} = \frac{-4}{2} = \mathbf{-2}$$

Horizontal and Vertical are same as above

**Positive slope**: As the x-value increases, the y-value also increases.
Analogy: Think of climbing a hill. As you move forward (increasing x), you go up (increasing y).

**Negative slope**: As the x-value increases, the y-value decreases.
Analogy: Think of sliding down a slide. As you move horizontally forward (increasing x), your height decreases (decreasing y).

### Moving left

The definition of positive and negative slope is based on the traditional reading direction: moving from left to right (where the x-value is increasing).

Lets take a world analogy, we start at time = 0 (x), distance travelled = 0 (y), after one hour time = 1, distance covered = +30 km, after 2 hours, distance is +60, after 3 hours distance travelled is 90km but between time 2 and 3 we moved downwards so it would be -30 km, but the total distance travelled is 90km, we can move up or down in y, but we can't move left, because from present we can't go to past 

Point A: $(x_1, y_1) = (1, 30)$ 

Point B: $(x_2, y_2) = (2, 60)$

Point C: $(x_3, y_3) = (3, 30)$ (Since you said between time 2 and 3, distance went from 60 down to $60 - 30 = 30$)

#### Falling
P1 = (4,4) and p2 = (1,3)
$$m = \frac{3-4}{1-4}=\frac{-1}{-3}=\frac{1}{3}=0.3$$

as per m, point moving upwords but actually point is moved downwards

#### Raising
p1 = (4,4) and p2 = (1, 5)

$$m = \frac{5-4}{1-4}=\frac{1}{-3}=\frac{1}{-3}=-0.3$$

as per m, point moving downwards but actually point is moved upwards

```python
# Quadrant Crossing Data Structure
# (Data copied from the previous table)
quad_crossing_data = {
    'Crossings (I to III & II to IV)': [ # Lines that cross the origin or span opposite quadrants
        {'P1': (1, 2), 'P2': (-3, 4), 'm': -1/2}, # I to II
        {'P1': (2, 1), 'P2': (-2, -1), 'm': 1/2}, # I to III (Crosses origin)
        {'P1': (3, 5), 'P2': (4, -2), 'm': -7},   # I to IV
        {'P1': (-3, 2), 'P2': (4, -4), 'm': -6/7}, # II to IV (Crosses axes)
        {'P1': (-4, 1), 'P2': (1, 3), 'm': 2/5},  # II to I
        {'P1': (-1, -3), 'P2': (3, 1), 'm': 1},   # III to I (Crosses origin)
    ],
    'Crossings (Around Y-axis)': [ # Lines crossing between II/III and I/IV
        {'P1': (-2, 5), 'P2': (-5, -1), 'm': 2},   # II to III
        {'P1': (-4, -2), 'P2': (-1, 3), 'm': 5/3}, # III to II
        {'P1': (1, -5), 'P2': (-2, 2), 'm': -7/3}, # IV to II
        {'P1': (5, -1), 'P2': (2, 3), 'm': -4/3},  # IV to I
    ],
    'Crossings (Around X-axis)': [ # Lines crossing between I/II and III/IV
        {'P1': (-5, -4), 'P2': (1, -3), 'm': 1/6}, # III to IV
        {'P1': (3, -2), 'P2': (-1, -4), 'm': 1/2}, # IV to III
    ]
}

# Create a single figure and subplot
fig, axes = plt.subplots(1, 3, figsize=(21, 7))
axes = axes.flatten()

# **GLOBAL LIMITS:** To encompass all points from -5 to 5
X_LIM_MIN, X_LIM_MAX = -6, 6
Y_LIM_MIN, Y_LIM_MAX = -6, 6

# Define a cycle of offsets for text labels to prevent overlap
offset_cycle = cycle([0.3, -0.3, 0.6, -0.6, 0.9, -0.9])

for i, (graph_name, lines) in enumerate(quad_crossing_data.items()):
    ax = axes[i]
    ax.set_title(f'Quadrant Crossing Slopes: {graph_name}', fontsize=14)
    ax.set_xlabel('$x$', fontsize=12)
    ax.set_ylabel('$y$', fontsize=12)
    ax.grid(True, linestyle='--', alpha=0.5)

    # Plot x and y axes to divide the quadrants
    ax.axhline(0, color='black', linewidth=1.5, zorder=0) # X-axis
    ax.axvline(0, color='black', linewidth=1.5, zorder=0) # Y-axis
    
    # Label the quadrants
    ax.text(X_LIM_MAX-0.5, Y_LIM_MAX-0.5, 'I (+,+)', fontsize=14, color='green', ha='right', va='top')
    ax.text(X_LIM_MIN+0.5, Y_LIM_MAX-0.5, 'II (-,+)', fontsize=14, color='blue', ha='left', va='top')
    ax.text(X_LIM_MIN+0.5, Y_LIM_MIN+0.5, 'III (-,-)', fontsize=14, color='red', ha='left', va='bottom')
    ax.text(X_LIM_MAX-0.5, Y_LIM_MIN+0.5, 'IV (+,-)', fontsize=14, color='purple', ha='right', va='bottom')

    # Apply limits and set ticks to scale 1
    ax.set_xlim(X_LIM_MIN, X_LIM_MAX)
    ax.set_ylim(Y_LIM_MIN, Y_LIM_MAX)
    ax.set_xticks(np.arange(X_LIM_MIN, X_LIM_MAX + 1, 1))
    ax.set_yticks(np.arange(Y_LIM_MIN, Y_LIM_MAX + 1, 1))
    ax.set_aspect('equal', adjustable='box')
    
    # Reset offset cycle for each graph
    offset_cycle = cycle([0.3, -0.3, 0.6, -0.6, 0.9, -0.9])

    for line_data in lines:
        x1, y1 = line_data['P1']
        x2, y2 = line_data['P2']
        m = line_data['m']

        # Get the next offset amount
        current_offset = next(offset_cycle)

        # Plot the line segment
        ax.plot([x1, x2], [y1, y2], marker='o', linestyle='-', linewidth=2, zorder=10)

        # Calculate midpoint
        mid_x = (x1 + x2) / 2
        mid_y = (y1 + y2) / 2

        # --- Angle Calculation (for rotation and offset) ---
        # Note: None of these lines are vertical or horizontal
        
        # Angle of the line segment
        angle_rad = np.arctan2(y2 - y1, x2 - x1)
        angle_deg = np.degrees(angle_rad)

        # Normalize angle
        if angle_deg > 90:
            angle_deg -= 180
        elif angle_deg < -90:
            angle_deg += 180

        # Calculate perpendicular angle for offset (add 90 degrees)
        perp_rad = angle_rad + np.pi/2
        # Apply offset along the perpendicular direction
        offset_x = current_offset * np.cos(perp_rad)
        offset_y = current_offset * np.sin(perp_rad)
        
        # --- Slope Text Formatting ---
        if m == 0:
            m_text = r'0'
        elif isinstance(m, float) and abs(m % 1) < 1e-9: # Check if close to an integer
             m_text = f'{int(m)}'
        elif m == 1/2:
            m_text = r'\frac{1}{2}'
        elif m == -1/2:
            m_text = r'-\frac{1}{2}'
        elif m == 2/5:
            m_text = r'\frac{2}{5}'
        elif m == 5/3:
            m_text = r'\frac{5}{3}'
        elif m == -6/7:
            m_text = r'-\frac{6}{7}'
        elif m == -7/3:
            m_text = r'-\frac{7}{3}'
        elif m == -4/3:
            m_text = r'-\frac{4}{3}'
        elif m == 1/6:
            m_text = r'\frac{1}{6}'
        else:
            m_text = f'{m}' # Use the original value for single integers like -7, 1, 2
        
        # Set final text coordinates by applying offset to the midpoint
        text_x = mid_x + offset_x
        text_y = mid_y + offset_y

        # Create the full label string
        label = (f'$({x1},{y1}) \\to ({x2},{y2})$\n$m={m_text}$')

        # Add the text label with rotation and offset
        ax.text(text_x, text_y, label, fontsize=9, color='darkred', weight='bold',
            ha='center', va='center', rotation=angle_deg,
            bbox=dict(facecolor='white', alpha=0.9, edgecolor='black', linewidth=0.5, boxstyle='round,pad=0.5', zorder=20))

plt.tight_layout(pad=3.0) 
```
