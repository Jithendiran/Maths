# Horizontal Expansion

$$\frac{\Delta x}{\Delta y} = \text{ Constant Rate of }\Delta x \text{ for single unit of }\Delta y$$

x denotes left or right, horizontal line, (x-axis)  

y denotes up or down, vertical line, (y-axis)

What we find here is for every 1 unit of y value, how much left or right the x value is

**Horizontal Expansion** is a term describes how the surface is expanded

**Horizontal Expansion** or **Invert Slope**

$$\text{ Invert Slope } = \frac{x2 - x1}{ y2 - y1} = \frac{\Delta x}{ \Delta y}$$

In algebra Invert Slope is denoted in $\frac{1}{m}$

if the horizontal line is small, line is very steep

if horizontal line very long, very little steep
