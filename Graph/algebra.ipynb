## What is a Graph?

A graph in the context of algebra is a visual representation of the relationship between two variables, typically drawn on a Coordinate Plane.

* It's typically drawn on a coordinate plane (like a grid), often called the Cartesian plane.

* The plane is defined by two perpendicular lines: the horizontal axis ($\boldsymbol{x}$-axis) and the vertical axis ($\boldsymbol{y}$-axis).

* Points on the graph are represented by ordered pairs $(\boldsymbol{x}, \boldsymbol{y})$, where the first number tells you the location along the $x$-axis and the second tells you the location along the $y$-axis.

### Coordinate Plane

The Coordinate Plane is a general term for any two-dimensional surface defined by a system that uses one or more coordinates (numbers) to uniquely determine the position of a point.

* **Generality**: It is the widespread concept. The axes do not necessarily have to be perpendicular (though they often are), nor do they have to be labeled $\boldsymbol{x}$ and $\boldsymbol{y}$.

* **Axes**: The axes represent independent and dependent variables (e.g., Time and Temperature, or Item Count and Cost). The starting point for the values (the origin) can be anywhere, typically at $(0, 0)$ or $(1, 1)$ in data graphing.

* **Use**: Often used in data visualization and statistics where the primary focus is mapping the relationship between two specific measured variables.

````python
import matplotlib.pyplot as plt
import numpy as np

# 1. Setup the figure and axes
fig, ax = plt.subplots(figsize=(7, 7))

# Define the range (starting from 1, common in basic data plots)
xmin, xmax = 0, 10
ymin, ymax = 0, 10
ax.set_xlim(xmin, xmax)
ax.set_ylim(ymin, ymax)
ax.set_aspect('equal', adjustable='box')

# 2. Draw the Grid Lines (The "Plane")
# Use a prominent grid across the entire viewing area
ax.grid(True, linestyle='-', linewidth=0.7, color='lightblue', alpha=0.8, zorder=0)

# 3. Add Labels
ax.set_xlabel('Independent Variable (e.g., Time)', fontsize=12)
ax.set_ylabel('Dependent Variable (e.g., Value)', fontsize=12)
ax.set_title('General Coordinate Plane / Data Grid', fontsize=14)

# 4. Plot a sample relationship (a simple scatter plot)
data_points = [(2, 3), (4, 6), (7, 8), (9, 5)]
x_data, y_data = zip(*data_points)
ax.scatter(x_data, y_data, color='darkgreen', s=100, zorder=3, label='Data Points')

# 5. Remove the thick center axes (spines) for a general look
ax.spines['left'].set_color('gray')
ax.spines['bottom'].set_color('gray')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.legend()
plt.show()
````

### Cartesian Plane

The Cartesian plane is a specific type of coordinate plane that uses the Cartesian coordinate system.

* Specificity: It is strictly defined by a few key rules:
1. Axes: It is defined by two perpendicular number lines: the horizontal $\boldsymbol{x}$-axis and the vertical $\boldsymbol{y}$-axis.
2. Origin: The axes must intersect at a specific point called the origin ($\boldsymbol{0, 0}$).
3. Coordinates: Points are located using the ordered pair $(\boldsymbol{x, y})$, where $x$ is the horizontal distance from the origin and $y$ is the vertical distance.
4. Quadrants: It is divided into four distinct Quadrants (I, II, III, IV) by the intersection of the axes.

#### Use

* It is particularly essential for visualizing a $\boldsymbol{\text{function}}$, which is a relationship where every input ($\boldsymbol{x}$) has exactly one output ($\boldsymbol{y}$).

intersect - The word intersect means to meet, cross, or overlap at a common point, line, or area.

perpendicular - Perpendicular is a geometric term describing two lines, segments, rays, or planes that intersect (cross) each other to form a perfect right angle, which measures exactly $90^\circ$

````python
import matplotlib.pyplot as plt
import numpy as np

# 1. Setup the figure and axes
fig, ax = plt.subplots(figsize=(7, 7))

# Define the range centered on zero
axis_range = 5
xmin, xmax = -axis_range, axis_range
ymin, ymax = -axis_range, axis_range
ax.set_xlim(xmin, xmax)
ax.set_ylim(ymin, ymax)
ax.set_aspect('equal', adjustable='box')

# 2. Draw the Grid Lines
ax.grid(True, which='major', linestyle=':', linewidth=0.5, color='gray', alpha=0.5, zorder=0)

# 3. Draw the Axes CORRECTLY using ax.annotate() for arrows
# X-Axis (Draws a line from the left edge to the right edge with an arrow)
ax.annotate('', xy=(xmax, 0), xytext=(xmin, 0), # Start at xmin, y=0 and end at xmax, y=0
            arrowprops={'arrowstyle': '->', 'lw': 2.0, 'color': 'black'}, 
            zorder=2)

# Y-Axis (Draws a line from the bottom edge to the top edge with an arrow)
ax.annotate('', xy=(0, ymax), xytext=(0, ymin), # Start at x=0, ymin and end at x=0, ymax
            arrowprops={'arrowstyle': '->', 'lw': 2.0, 'color': 'black'}, 
            zorder=2)


# 4. Set Labels and Title
ax.set_xlabel('X-Axis', fontsize=12)
ax.set_ylabel('Y-Axis', fontsize=12, rotation=0, labelpad=15)
ax.set_title('The Cartesian Plane (Defining Quadrants)', fontsize=14)

# 5. Label the Origin (0, 0) and Quadrants
ax.text(0.1, -0.4, 'O (0, 0)', fontsize=12, color='black')
ax.text(3, 4, 'I (+, +)', fontsize=12, color='blue')
ax.text(-4.8, 4, 'II (-, +)', fontsize=12, color='blue')
ax.text(-4.8, -4.8, 'III (-, -)', fontsize=12, color='blue')
ax.text(3, -4.8, 'IV (+, -)', fontsize=12, color='blue')

# 6. Plot a sample point in Quadrant II
point_x, point_y = -3, 2
ax.plot(point_x, point_y, 'go', markersize=8, zorder=3, label=f'Point ({point_x}, {point_y})')

plt.legend()
plt.show()
````

## Why Do We Need Graphs in Algebra?
Graphs provide a powerful, intuitive way to understand algebraic concepts.

1. Visualization
Graphs help you see the solutions and behavior of equations. An equation like $y = 2x + 1$ can be represented by a straight line. Every point on that line is a pair $(x, y)$ that makes the equation true.

2. Understanding Relationships
They show the relationship between variables. For example, if you graph the distance a car travels over time, the graph's steepness (slope) tells you the car's speed.

3. Solving Systems of Equations
When you have two or more equations (a system), the point(s) where their graphs intersect is the solution to the system. This is often easier to grasp visually than with just algebraic manipulation.

### Linear Equations and Line

* A linear equation is an algebraic equation that, when plotted on the Cartesian plane, always results in a straight **line**.
* The key characteristic of a linear equation is that the variables (usually $x$ and $y$) are raised only to the first power;
* they are never multiplied together, nor do they appear in exponents, roots.

Most of the time we treat $x$ as power one, we find the $y$ with respect to $x$, so $x$ is independent variable and $y$ is dependent variable

#### example

##### Linear

$$3x+2y=6$$
$$\text{x == } x^1 \text{, y == } y^1 \text{, all variables are raised to power of 1}$$
$$\text{x and y are not multiplied together}$$
$$\text{There is no square, roots, exponents in the equation}$$

#### Non linear
* $$x^2 + y = 4$$
* $$xy = 5$$

###  Forms of a Linear Equation

#### Solving for variables

* Single variable
$$Ax+B=C \text{ A, B and C are constant, x is variable, A is also a coefficient}$$
$$x=\frac{(C-B)}{A}$$

* Two variables
$$Ax + By = C$$
$$\text{A and B are coefficient, x and y are variables, C is a constant}$$

$$x=\frac{C-By}{A}, y=\frac{C-Ax}{B}$$

* n variable
$$Ax + By + Cz  = D$$


| **Linear Equation Form** | **General Form**                  | **Example**                       |
| ------------------------ | --------------------------------- | --------------------------------- |
| **SlopeIntercept Form** | $y = mx + b$                    | $y + 2x = 3$                    |
| **PointSlope Form**     | $y - y_1 = m(x - x_1)$          | $y - 3 = 6(x - 2)$              |

#### 1. Constant Function (Vertical line)

$$f(x) = C \text{ OR } y = C$$

C  is the constant

$f(x) = C$ is Function notation

$f(x)$ represents the output value of the function (which is the same as $y$) and is always equal to the constant $C$.

When you see $f(x)=C$, it specifically tells you that for any input value of $x$ (e.g., $f(1), f(10), f(-50)$), the output always equals $C$

$f(x) = 3$ means $f(1) = 3$, $f(100) = 3$, and $f(\text{anything}) = 3$.

$y = C$ is Equation of a line

##### How to Solve/Graph the Equation

To graph $x = a$ (e.g., $x = -2$):

Understand the Rule: The rule says that the $x$-value must be $-2$, regardless of what $y$-value you choose.

If $y=0$, $x=-2 \implies (-2, 0)$

If $y=4$, $x=-2 \implies (-2, 4)$

````python
import matplotlib.pyplot as plt
import numpy as np

# 1. Define the constant 'x'
x = -2

# 2. Define a range of y-values for the plot
# For a vertical line, y is the independent variable in the plot setup
y = np.linspace(-5, 5, 100)

# 3. Calculate the corresponding x-values

a = np.full_like(y, x) # x is an array where every value is -2

# 4. Create the plot
plt.figure(figsize=(8, 6))

# Plot the x line 
plt.plot(a, y, label=f'x = {x}', color='red')

# Add a point
plt.plot(x, 0, 'go', label=f'({x}, 0)')
plt.plot(x, -2, 'go', label=f'({x}, -2)')
plt.plot(x, 4, 'go', label=f'({x}, 4)')
plt.plot(x, 1, 'go', label=f'({x}, 1)')

# 5. Add labels and title
plt.title(f'Graph of a Vertical Line: x = {x} (NOT a function)')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True)

# 6. Ensure the axes pass through the origin (0,0)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Set the limits to make the vertical nature clear
plt.xlim(-5, 5)
plt.ylim(-5, 5)

# Display the plot
plt.show()
````

#### 2. The Identity Function (Diagonal Line)

Equation: $\mathbf{y = x}$ or $\mathbf{f(x) = x}$

Identity Function: In mathematics, the "identity" property means that when you perform an operation, the value remains unchanged. Here, the output identically matches the input: $y$ is the identical value to $x$.

##### How to Solve/Graph the Equation
Understand the Rule: The $y$-coordinate is always the same as the $x$-coordinate.

If $x=0$, $y=0 \implies (\mathbf{0}, \mathbf{0})$

If $x=4$, $y=4 \implies (\mathbf{4}, \mathbf{4})$

````python
import matplotlib.pyplot as plt
import numpy as np

# 1. Define the range of x-values
x = np.linspace(-5, 5, 100)

# 2. Calculate the corresponding y-values
# For the Identity Function, y is simply equal to x
y = x

# 3. Create the plot
plt.figure(figsize=(8, 6))

# Plot the line
plt.plot(x, y, label='y = x (Identity Function)', color='purple', linewidth=2)

# Optional: Highlight the origin (0, 0)
plt.plot(0, 0, 'go', label='(0, 0)')
plt.plot(4, 4, 'go', label='(0, 0)')
plt.plot(-4, -4, 'go', label='(0, 0)')

# 4. Add labels and title
plt.title('Graph of the Identity Function: y = x')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True)

# 5. Ensure the axes pass through the origin (0,0) and set equal scaling
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box') # Makes 1 unit on x-axis equal to 1 unit on y-axis, ensuring a 45-degree angle.

# Display the plot
plt.show()
````

### 3. Parallel and Perpendicular Lines

**Parallel Lines:**
* Two lines are parallel if they never intersect, meaning they have the same steepness and rise at the same rate.
* Two non-vertical lines are parallel if and only if they have the same slope.
* Any line parallel to the segment between $(1, 1)$ and $(3, 3)$ must also have a slope of $m_{\parallel} = 1$

**Perpendicular Lines:**
* Two non-vertical lines are perpendicular if the product of their slopes is $-1$. This means the slope of the perpendicular line is the negative reciprocal of the original line's slope.
* Formula: $m_{\perp} = -\frac{1}{m}$

#### 6. Intercept Form

What is an Intercept?

An intercept is a point where a line (or any graph) crosses or touches one of the coordinate axes.

Since the coordinate plane has two main axes (the $x$-axis and the $y$-axis), a line can have two types of intercepts

1. The $x$-Intercept (Horizontal Intercept): 
* Definition: The point where the line crosses the horizontal axis (the $x$-axis).
* Key Feature: At this point, the vertical distance from the axis is zero Therefore, the $y$-coordinate must be 0.
* Form: An $x$-intercept always has the form $\mathbf{(\text{number}, 0)}$, such as $(5, 0)$ or $(-3, 0)$.

2. The $y$-Intercept (Vertical Intercept): 
* Definition: The point where the line crosses the vertical axis (the $y$-axis).
* Key Feature: At this point, the horizontal distance from the axis is zero. Therefore, the $x$-coordinate must be 0.
* Form: An $y$-intercept always has the form $\mathbf{(0, \text{number})}$, such as $(0, 4)$ or $(0, -1)$

Standard form
-----

$$\frac{x}{a} + \frac{y}{b} = 1$$

Why does the formula $\frac{x}{a} + \frac{y}{b} = 1$ work? It's structured so that when one variable is zero, the other must equal its intercept value.

| To find the... | You set... | The equation becomes... | Which solves to... |
| :---: | :---: | :---: | :---: |
| **$x$-intercept** | $\mathbf{y = 0}$ | $\frac{x}{a} + \frac{0}{b} = 1 \implies \frac{x}{a} = 1$ | **$x = a$** |
| **$y$-intercept** | $\mathbf{x = 0}$ | $\frac{0}{a} + \frac{y}{b} = 1 \implies \frac{y}{b} = 1$ | **$y = b$** |

Consider this equation in Intercept Form:
$$\frac{x}{5} + \frac{y}{-2} = 1$$

* The denominator under $x$ is $a=5$. The $x$-intercept is $(\mathbf{5}, 0)$.
* The denominator under $y$ is $b=-2$. The $y$-intercept is $(\mathbf{0}, -2)$.

Let's convert the **Standard Form** equation $2x + 3y = 6$ into Intercept Form.

The goal is to make the right side of the equation equal to 1.

1.  **Start with the Standard Form:**
    $$2x + 3y = 6$$
2.  **Divide the entire equation by the constant (6):**
    $$\frac{2x}{6} + \frac{3y}{6} = \frac{6}{6}$$
3.  **Simplify the fractions:**
    $$\frac{x}{3} + \frac{y}{2} = 1$$

Intercept form
----

In the **Slope-Intercept Form** ($y = mx + \mathbf{c}$):
* The letter $\mathbf{c}$ is the **value of the $y$-intercept**.
* It tells you that the line crosses the $y$-axis at the point $(\mathbf{0, c})$.

| To Find | Rule | Example: For $y = 2x + 6$ |
| :---: | :--- | :--- |
| **$y$-intercept** | Set $\mathbf{x = 0}$ | $y = 2(0) + 6 \implies y = 6$. The $y$-intercept is $\mathbf{(0, 6)}$. |
| **$x$-intercept** | Set $\mathbf{y = 0}$ | $0 = 2x + 6 \implies -6 = 2x \implies x = -3$. The $x$-intercept is $\mathbf{(-3, 0)}$. |

````python
# Define the equation and range
# y = 2x + 6
x = np.linspace(-5, 5, 100)
y = 2 * x + 6

# Define the intercepts for highlighting
x_intercept = -3
y_intercept = 6

# Create the plot
plt.figure(figsize=(8, 6))

# Plot the line
plt.plot(x, y, label='$y = 2x + 6$', color='blue', linewidth=2)

# Highlight the intercepts
plt.scatter(x_intercept, 0, color='red', s=100, zorder=5, label=f'x-intercept ({x_intercept}, 0)')
plt.scatter(0, y_intercept, color='green', s=100, zorder=5, label=f'y-intercept (0, {y_intercept})')

# Add labels, title, and grid
plt.title('Graph of the Linear Equation $y = 2x + 6$', fontsize=16)
plt.xlabel('x-axis', fontsize=12)
plt.ylabel('y-axis', fontsize=12)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Set limits
plt.xlim(-5, 5)
plt.ylim(-5, 15)
````
